```python
import pandas as pd
import numpy as np
import random

# Number of rows to generate
n_rows = 100

# Random fruits
fruits = random.choices(["Apple", "Pear", "Cherry"], k=n_rows)

# Random years and months
years = random.choices([2022, 2023], k=n_rows)
months = random.choices(np.arange(1,3), k=n_rows)

# Random prices and volumes
weights = np.random.rand(n_rows)
volumes = np.random.rand(n_rows)

# Create DataFrame
df = pd.DataFrame({
    'Fruit': fruits,
    'Year': years,
    'Month': months,
    'Weight': weights,
    'Volume': volumes
})

# Perform groupby, sum and unstack operations
df_grouped = df.groupby(['Fruit', 'Year', 'Month']).sum().unstack('Fruit').swaplevel(0, 1, axis=1).sort_index(axis=1)

df_grouped
```
